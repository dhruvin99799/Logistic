# IQR

# 8].Finding the IQR of ApplicantIncome

```python
per25 = df['ApplicantIncome'].quantile(0.25)
per75 = df['ApplicantIncome'].quantile(0.75)
```

```python
print(per25)
print(per75)
```

```python
iqr = per75 - per25
iqr
```

```python
upper_limit = per75 + 1.5 * iqr
lower_limit = per25 - 1.5 * iqr
```

```python
print(upper_limit)
print(lower_limit)
```

```python
# Finding Outliers In ApplicantIncome
```

```python
df[df['ApplicantIncome'] > upper_limit]
```

```python
df[df['ApplicantIncome'] < lower_limit]
```

```python
# Removing to data With Trim Method
```

```python
trim_df =df[df['ApplicantIncome'] < upper_limit]
trim_df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['ApplicantIncome'])
plt.gca().set_facecolor('#274c77')


plt.subplot(2,2,2)
sns.boxplot(df['ApplicantIncome'])
plt.gca().set_facecolor('#274c77')


plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(trim_df['ApplicantIncome'])
plt.gca().set_facecolor('#274c77')


plt.subplot(2,2,2)
sns.boxplot(trim_df['ApplicantIncome'])
plt.gca().set_facecolor('#274c77')


plt.show()
```

```python
# Removing to data With Cap Method
```

```python
cap_df = df.copy()
cap_df['ApplicantIncome'] = np.where(
    cap_df['ApplicantIncome'] > upper_limit,
    upper_limit,
    np.where(
        cap_df['ApplicantIncome'] < lower_limit,
        lower_limit,
        cap_df['ApplicantIncome']
    )
)
```

```python
cap_df.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
plt.gca().set_facecolor('#274c77')
sns.distplot(df['ApplicantIncome'])

plt.subplot(2,2,2)
plt.gca().set_facecolor('#274c77')
sns.boxplot(df['ApplicantIncome'])

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
plt.gca().set_facecolor('#274c77')
sns.distplot(cap_df['ApplicantIncome'])

plt.subplot(2,2,2)
plt.gca().set_facecolor('#274c77')
sns.boxplot(cap_df['ApplicantIncome'])

plt.show()
```

# Logistical Regration

```python
X=df[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']]
y=df['Loan_Status']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
sc = StandardScaler()
X[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']] = sc.fit_transform(X[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']])
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
one = OneHotEncoder()
one.fit(X_train[['Gender', 'Married', 'Dependents','Education','Self_Employed','Property_Area']])
```

```python
col_tr = make_column_transformer((OneHotEncoder(categories=one.categories_),['Gender', 'Married', 'Dependents', 'Education','Self_Employed','Property_Area']),remainder='passthrough')
col_tr
```

```python
classifer = LogisticRegression()
```

```python
pipe = make_pipeline(col_tr,classifer)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```
